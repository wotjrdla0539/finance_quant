```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
S0 = 30
n = 52*5
mu = 0.09
sigma = 0.20
e = (mu/52 + sigma*np.random.randn(n, 100 ) / math.sqrt(52) + 1)
S = S0 * e.cumprod(axis=0)
# vstack -  
S = np.vstack([np.ones((1, 100))*S0, S])
```

```python
plt.plot(S)
```
